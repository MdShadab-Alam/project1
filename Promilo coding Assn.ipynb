# Top 5 Highest Page path and screen class based on Average engagement time (Pages_and_Screens_Report data)

```python
highest_Page_path_and_screen_class_by_Average_engagement_time=Pages_and_Screens_Report.sort_values(by=['Average engagement time'],ascending=False)[['Average engagement time','Page path and screen class']].head(5)
```

```python
data = highest_Page_path_and_screen_class_by_Average_engagement_time
x = highest_Page_path_and_screen_class_by_Average_engagement_time['Page path and screen class']
y = highest_Page_path_and_screen_class_by_Average_engagement_time['Average engagement time']

ax = sns.lineplot(data=data, x=x, y=y)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

for x_pos, y_pos in zip(x, y):
    plt.annotate(f'{y_pos:.2f}', (x_pos, y_pos), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
